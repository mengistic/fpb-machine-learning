# Problem 1

Consider the function $f(x)=\cos(x^3+\sin(x))$ defined on $\mathbb R$. 

1. Graph the function $f$ on the interval $[-5,5]$
2. Graph the function $f(f(x))$ on $[-5,5]$
3. Find the integral $I=\int_0^1\cos(x^3+\sin(x))\,dx$.

---


## Solution:

```python
import matplotlib.pyplot as plt 
import numpy as np 
import scipy as sci
import sympy as sym

def f(x):
    return np.cos(x**3+np.sin(x))
```

```python
## Graphing the functions 
x = np.linspace(-5,5,500) 

plt.subplot(2,1,1) 
plt.plot(x, f(x))
plt.title("Graph of $f(x)$")

plt.subplot(2,1,2)
plt.plot(x,f(f(x)))
plt.title("Graph of $f(f(x))$")
plt.subplots_adjust(top=1.5, bottom=0.1, left=0.1, right=1, hspace=0.25, wspace=0.5)
```

## Finding the integral
To evaluate the integral, I use the function `integrate.quad` from `scipy` library. It returns a tuple whose first index represents the value of integration and the second one represent the maximum error.

```python
I = sci.integrate.quad(f, -5,5)
I
```
